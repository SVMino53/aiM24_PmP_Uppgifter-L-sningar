```python
import matplotlib.pyplot as plt
import numpy as np
import random


fig = plt.figure()

ax : plt.Axes = fig.add_axes([0, 0, 0.9, 0.9])
x_vals = list(np.linspace(-10*np.pi, 10*np.pi, 1000))
y_vals = [np.sin(x) for x in x_vals]
ax.plot(x_vals, y_vals)

ax : plt.Axes = fig.add_axes([1, 0, 0.9, 0.9])
x_vals = list(np.linspace(0, 10, 1000))
y_vals = [np.e**x for x in x_vals]
ax.plot(x_vals, y_vals)

ax : plt.Axes = fig.add_axes([0, -1, 0.9, 0.9])
x_vals = list(np.linspace(0, 10))
y_vals = [x + x*random.random() for x in x_vals]
ax.scatter(x_vals, y_vals)

ax : plt.Axes = fig.add_axes([1, -1, 0.9, 0.9])
x_vals = list(np.linspace(0, 10))
vals = [sum([random.randint(1, 6) for rolls in range(4)]) for _ in range(1000)]
ax.hist(vals, bins=[4 + x for x in range(24)], linewidth=0.5, edgecolor="white")


plt.show()
```
